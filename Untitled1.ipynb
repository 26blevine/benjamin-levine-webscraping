```python
# cell 1: imports and driver creation
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager   # installs driver automatically
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import os

# 1) Chrome options (no headless — this is "headful")
options = webdriver.ChromeOptions()
# optional: start maximized so you can see it
options.add_argument("--start-maximized")
# optional: prevent automation infobar (may or may not work depending on Chrome version)
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option('useAutomationExtension', False)

# 2) Create driver using webdriver-manager to auto-download the right chromedriver
service = ChromeService(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# 3) Example navigation + explicit wait (replace URL and selector as needed)
url = "https://www.bhphotovideo.com/c/buy/Mirrorless-Camera-Lenses/ci/17912/N/4196380428"
driver.get(url)

# Wait up to 15s for <h1> to be present, then print the page title
wait = WebDriverWait(driver, 15)
h1 = wait.until(EC.presence_of_element_located((By.TAG_NAME, "h1")))
print("Page title:", driver.title)
print("h1 text:", h1.text)

# 4) Screenshot (saved into notebook's working directory)
screenshot_path = "selenium_visible_screenshot.png"
driver.save_screenshot(screenshot_path)
print("Saved screenshot to", os.path.abspath(screenshot_path))

# 5) Keep the browser open a bit so you can see actions (optional)
time.sleep(150)

# 6) Close when you're done
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
import time

# Chrome setup — visible (non-headless) mode
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")

# Start the driver
service = ChromeService(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)
driver.execute_cdp_cmd("Emulation.setScriptExecutionDisabled", {"value": True})

# Target page
url = "https://www.bhphotovideo.com/c/buy/Mirrorless-Camera-Lenses/ci/17912/N/4196380428"
driver.get(url)

# === Immediately save the full HTML ===
html_source = driver.page_source   # grab the rendered HTML
with open("page_source.txt", "w", encoding="utf-8") as f:
    f.write(html_source)
print("Saved HTML to page_source.txt")

# Optional: keep window open a few seconds
time.sleep(15)

# Close the browser
driver.quit()
```
